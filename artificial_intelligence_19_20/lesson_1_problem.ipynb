```python
def BFS_TreeSearch(problem):
    """
    Tree Search BFS

    Args:
        problem: OpenAI Gym environment

    Returns:
        (path, time_cost, space_cost): solution as a path and stats.
    """

    # The node with STATE = problem.INITIAL-STATE.
    node = Node(problem.startstate, None)
    
    time_cost = 0
    space_cost = 0

    # Check if node.STATE is a problem.GOAL.
    if (node.state == problem.goalstate):
        return build_path(node), time_cost, space_cost

    # Create and initialize the frontier.
    frontier = NodeQueue()
    frontier.add(node)
    
    space_cost += 1

    while True:
        # Check if EMPTY?(frontier).
        if frontier.is_empty():
            return None, 0, 0

        # node <- POP(frontier). Chooses the shallowest node in frontier (FIFO).
        tmp_node = frontier.remove()
        space_cost -= 1

        # For each action in problem.ACTIONS(node.STATE).
        for action in range(problem.action_space.n):
            child_state_id = problem.sample(tmp_node.state, action)
            child_state_node = Node(child_state_id, tmp_node)

            # Check if child_state_node.STATE is a problem.GOAL.
            if child_state_node.state == problem.goalstate:
                return build_path(child_state_node), time_cost, space_cost

            # frontier <- INSERT(child_state_node, frontier).
            frontier.add(child_state_node)
            time_cost += 1
            space_cost += 1
```

```python
def BFS_GraphSearch(problem):
    """
    Graph Search BFS

    Args:
        problem: OpenAI Gym environment

    Returns:
        (path, time_cost, space_cost): solution as a path and stats.
    """

    # The node with STATE = problem.INITIAL-STATE.
    node = Node(problem.startstate, None)
    
    time_cost = 0
    space_cost = 0

    # Check if node.STATE is a problem.GOAL.
    if (node.state == problem.goalstate):
        return build_path(node), time_cost, space_cost

    # Create and initialize the frontier.
    frontier = NodeQueue()
    frontier.add(node)

    # An empty set for child already explored.
    explored = set()

    while True:
        # Check if EMPTY?(frontier).
        if frontier.is_empty():
            return None, 0, 0

        # node <- POP(frontier). Chooses the shallowest node in frontier (FIFO).
        tmp_node = frontier.remove()
        
        # Add node.STATE to explored.
        explored.add(tmp_node.state)
        space_cost += 1

        # For each action in problem.ACTIONS(node.STATE).
        for action in range(problem.action_space.n):
            child_state_id = problem.sample(tmp_node.state, action)
            child_state_node = Node(child_state_id, tmp_node)

            # Check if child.STATE is not in explored or frontier.
            if (child_state_node.state not in explored) and (child_state_node.state not in frontier):
                
                # Check if child_state_node.STATE is a problem.GOAL.
                if child_state_node.state == problem.goalstate:
                    return build_path(child_state_node), time_cost, space_cost

                # Push the child_state_node in the frontier.
                frontier.add(child_state_node)

            time_cost += 1
```

```python
def DLS(problem, limit, RDLS_Function):
    """
    DLS
    
    Args:
        problem: OpenAI Gym environment
        limit: depth limit for the exploration, negative number means 'no limit'
        
    Returns:
        (path, time_cost, space_cost): solution as a path and stats.
    """
        
    node = Node(problem.startstate, None)
    return RDLS_Function(node, problem, limit, set())
```

The next three functions have to be implemented:

```python
def Recursive_DLS_GraphSearch(node, problem, limit, closed):
    """
    Recursive DLS

    Args:
        node: node to explore
        problem: OpenAI Gym environment
        limit: depth limit for the exploration, negative number means 'no limit'
        closed: completely explored nodes

    Returns:
        (path, time_cost, space_cost): solution as a path and stats.
    """

    # Check if node.STATE is a problem.GOAL.
    if node.state == problem.goalstate:
        return build_path(node), 1, node.pathcost
    # Else, if limit = 0 then return cutoff
    elif limit == 0:
        return "cut_off", 1, node.pathcost
    else:
        # Add node.STATE to closed.
        closed.add(node.state)

        space_cost = node.pathcost
        time_cost = 1

        # cutoff_occurred <- false.
        cutoff_occurred = False

        # For each action in problem.ACTIONS(node.STATE).
        for action in range(problem.action_space.n):
            child_state_id = problem.sample(node.state, action)
            # Check if child.STATE is not in closed (completely explored nodes).
            if child_state_id not in closed:
                child_state_node = Node(child_state_id, node, node.pathcost + 1)
                
                # result <- RECURSIVE-DLS(child, problem, limit - 1, closed)
                result = Recursive_DLS_GraphSearch(child_state_node, 
                                                   problem, 
                                                   limit - 1, 
                                                   closed)

                time_cost += result[1]

                if result[2] > space_cost:
                    space_cost = result[2]

                if result[0] == "cut_off":
                    cutoff_occurred = True
                elif result[0] != "failure":
                    return result[0], time_cost, space_cost

        # If cutoff_occurred? (not all states have been explored).
        if cutoff_occurred:
            return "cut_off", time_cost, space_cost
        else:
            return "failure", time_cost, space_cost
```

```python
def Recursive_DLS_TreeSearch(node, problem, limit, closed=None):
    """
    DLS (Tree Search Version)

    Args:
        node: node to explore
        problem: OpenAI Gym environment
        limit: depth limit for the exploration, negative number means 'no limit'

    Returns:
        (path, time_cost, space_cost): solution as a path and stats.
    """

    # Check if node.STATE is a problem.GOAL.
    if node.state == problem.goalstate:
        return build_path(node), 1, node.pathcost
    # Else, if limit = 0 then return cutoff
    elif limit == 0:
        return "cut_off", 1, node.pathcost
    else:
        space_cost = node.pathcost
        time_cost = 1

        # cutoff_occurred <- false.
        cutoff_occurred = False

        # For each action in problem.ACTIONS(node.STATE).
        for action in range(problem.action_space.n):
            child_state_id = problem.sample(node.state, action)
            # child <- CHILD-NODE(problem, node, action)
            child_state_node = Node(child_state_id, node, node.pathcost + 1)

            # result <- RECURSIVE-DLS(child, problem, limit - 1)
            result = Recursive_DLS_TreeSearch(child_state_node, 
                                              problem, 
                                              limit - 1)

            time_cost += result[1]
            
            if result[2] > space_cost:
                space_cost = result[2]

            if result[0] == "cut_off":
                cutoff_occurred = True
            elif result[0] != "failure":
                return result[0], time_cost, space_cost

        # If cutoff_occurred? (not all states have been explored).
        if cutoff_occurred:
            return "cut_off", time_cost, space_cost
        else:
            return "failure", time_cost, space_cost
```
